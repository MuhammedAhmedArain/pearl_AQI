# Exploratory Data Analysis (EDA) - Pearls AQI Predictor

This notebook performs EDA to identify trends in Air Quality Index across time and weather conditions, as required by the internship guidelines.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sys
from pathlib import Path
sys.path.insert(0, str(Path().resolve().parent))
import config

plt.style.use('dark_background')
sns.set_palette("viridis")
```

### Load Data

```python
df = pd.read_csv(config.RAW_DATA_FILE, parse_dates=['timestamp'])
df.head()
```

### 1. AQI Distribution

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['aqi'], bins=30, kde=True)
plt.title('Distribution of AQI')
plt.xlabel('Air Quality Index')
plt.show()
```

### 2. Time-Series Trend of AQI

```python
df_time = df.set_index('timestamp')
df_time['aqi'].rolling(window=24).mean().plot(figsize=(14, 5))
plt.title('24-Hour Rolling Average of AQI Over Time')
plt.ylabel('AQI')
plt.show()
```

### 3. Correlation Heatmap (Weather vs Pollutants)

```python
numeric_df = df.select_dtypes(include=['float64', 'int64'])
plt.figure(figsize=(12, 8))
sns.heatmap(numeric_df.corr(), annot=False, cmap='coolwarm', center=0)
plt.title('Correlation Matrix')
plt.show()
```
